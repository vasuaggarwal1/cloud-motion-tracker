```python
import h5py
import numpy as np
import os
from datetime import datetime

# =====================================================
# CONFIGURATION
# =====================================================
INPUT_DIR  = r"C:\Users\parth\OneDrive\Desktop\dataset\INSET 3DR dataset\Dataset\Feb26_171204"
OUTPUT_DIR = r"C:\Users\parth\OneDrive\Desktop\dataset\INSET 3DR dataset\Dataset\cmk"
os.makedirs(OUTPUT_DIR, exist_ok=True)

LAT_MIN, LAT_MAX = 5.0, 35.0
LON_MIN, LON_MAX = 65.0, 100.0
MIN_CLOUD_FRACTION = 0.15

# =====================================================
# HELPERS
# =====================================================

def read_latlon(f):
    """Read, scale and mask Latitude / Longitude from an open HDF5 file."""
    lat = f["Latitude"][:].astype(np.float32)
    lon = f["Longitude"][:].astype(np.float32)
    # INSAT-3DR stores lat/lon as int16 scaled by 100; fill = 32767
    lat[lat == 32767] = np.nan
    lon[lon == 32767] = np.nan
    lat /= 100.0
    lon /= 100.0
    return lat, lon


def compute_india_bbox(lat, lon):
    """
    Compute the row/col bounding box of the Indian subcontinent crop.

    FIX vs. original: longitude columns are now determined using a 2D mask
    (lat AND lon conditions applied together) instead of a single reference
    row.  The single-row approach was fragile because:
      - In a geostationary projection the lon<->col mapping varies with row.
      - If the reference row happened to have NaNs it silently failed.

    Returns
    -------
    rmin, rmax : int   row indices (in lat/lon grid space)
    cmin, cmax : int   col indices (in lat/lon grid space)
    """
    # ---- Latitude crop (row range) ----
    lat_mask = np.isfinite(lat) & (lat >= LAT_MIN) & (lat <= LAT_MAX)
    lat_rows = np.where(np.any(lat_mask, axis=1))[0]
    if lat_rows.size == 0:
        raise ValueError("No rows found in the latitude range [%.1f, %.1f]" %
                         (LAT_MIN, LAT_MAX))
    rmin, rmax = int(lat_rows.min()), int(lat_rows.max())

    # ---- Longitude crop (col range) — 2D mask approach ----
    #
    # FIX: Build a 2D mask combining both lat and lon constraints.
    # np.any(..., axis=0) then projects this down to a per-column flag,
    # giving a column range that covers *all* India pixels regardless of row.
    # This is robust to curvilinear grids and avoids the single-ref-row bug.
    lon_mask  = np.isfinite(lon) & (lon >= LON_MIN) & (lon <= LON_MAX)
    combined  = lat_mask & lon_mask          # pixels that are BOTH in lat & lon range
    col_flags = np.any(combined, axis=0)     # True for every col that has ≥1 India pixel
    india_cols = np.where(col_flags)[0]

    if india_cols.size == 0:
        raise ValueError(
            "Could not locate India columns for lon range [%.1f, %.1f]" %
            (LON_MIN, LON_MAX))

    cmin, cmax = int(india_cols.min()), int(india_cols.max())
    return rmin, rmax, cmin, cmax


def _get_spatial_dims(cmk_ds):
    """
    Return the two *spatial* dimension indices of the CMK dataset.

    FIX: INSAT-3DR stores CMK as a 3-D array (1, rows, cols) — a leading
    singleton band dimension.  The original code blindly used cmk_ds.shape[0]
    and cmk_ds.shape[1] as rows and cols, which made row_scale ≈ 1/2816 and
    treated band-index slices as spatial-row slices.  That completely broke
    the column crop (the third dimension was returned in full).

    This helper detects the band dimension and returns the indices of the two
    genuine spatial dimensions so the rest of read_cmk_crop can work
    correctly for both 2-D and 3-D storage layouts.

    Returns
    -------
    row_dim, col_dim : int   dimension indices for rows and columns
    """
    ndim = cmk_ds.ndim
    if ndim == 2:
        return 0, 1
    elif ndim == 3:
        # Singleton dimensions are band/time dimensions; spatial dims are the
        # two largest.
        sizes = list(cmk_ds.shape)
        sorted_dims = sorted(range(ndim), key=lambda d: sizes[d], reverse=True)
        row_dim, col_dim = sorted(sorted_dims[:2])   # keep row < col order
        return row_dim, col_dim
    else:
        raise ValueError(f"Unexpected CMK dimensionality: {ndim}D {cmk_ds.shape}")


def read_cmk_crop(f, rmin, rmax, cmin, cmax, ll_shape):
    """
    Read the CMK crop, automatically:
      * handling 3-D datasets with a leading band dimension, AND
      * rescaling indices if CMK has a different spatial resolution than
        the Latitude/Longitude arrays.

    Returns a plain numpy uint8 array (fill pixels set to 0).

    FIX summary
    -----------
    The original code used cmk_shape[0]/cmk_shape[1] directly as the
    spatial row/col sizes.  When CMK is stored as (1, rows, cols) that made:
        row_scale = 1 / ll_rows  ≈ 0   → r0=0, r1=1  (selected band 0 only)
        col_scale = rows / ll_cols ≈ 1  → c0≈cmin, c1≈cmax (applied to dim-1
                                           which is the ROW dimension!)
    h5py's 2-D slice cmk_ds[0:1, cmin:cmax] then returned the full third
    dimension (2805 columns) untouched — so the column crop was silently
    skipped every time.

    The fix: identify spatial dimensions properly, then slice all three axes
    explicitly when the dataset is 3-D.
    """
    cmk_ds   = f["CMK"]
    row_dim, col_dim = _get_spatial_dims(cmk_ds)

    # Spatial shape only
    sp_rows  = cmk_ds.shape[row_dim]
    sp_cols  = cmk_ds.shape[col_dim]
    ll_rows, ll_cols = ll_shape

    # ---- scale bbox to CMK resolution (handles multi-res files) ----
    row_scale = sp_rows / ll_rows
    col_scale = sp_cols / ll_cols

    r0 = int(np.floor(rmin * row_scale))
    r1 = int(np.ceil( rmax * row_scale)) + 1
    c0 = int(np.floor(cmin * col_scale))
    c1 = int(np.ceil( cmax * col_scale)) + 1

    # Clamp to dataset bounds
    r0 = max(r0, 0);  r1 = min(r1, sp_rows)
    c0 = max(c0, 0);  c1 = min(c1, sp_cols)

    if r0 >= r1 or c0 >= c1:
        raise ValueError(
            f"Scaled CMK crop [{r0}:{r1}, {c0}:{c1}] is empty. "
            f"Spatial shape=({sp_rows},{sp_cols}), Lat/Lon shape={ll_shape}"
        )

    # ---- Read — handle 2-D and 3-D layouts ----
    if cmk_ds.ndim == 2:
        raw = cmk_ds[r0:r1, c0:c1]
    else:
        # 3-D: build an index tuple so we crop ALL three axes explicitly.
        # The non-spatial (band/time) dimensions collapse to full slices.
        idx = [slice(None)] * cmk_ds.ndim
        idx[row_dim] = slice(r0, r1)
        idx[col_dim] = slice(c0, c1)
        raw = cmk_ds[tuple(idx)]
        # Squeeze out all non-spatial leading/trailing singleton dims
        raw = np.squeeze(raw)

    # h5py may return a masked array when a _FillValue is set
    if isinstance(raw, np.ma.MaskedArray):
        fill_val = raw.fill_value
        data = raw.filled(0).astype(np.uint8)
        data[raw.data == fill_val] = 0
    else:
        data = np.array(raw, dtype=np.uint8)
        data[(data == 255)] = 0  # typical INSAT-3DR CMK fill

    return data, (r0, r1-1, c0, c1-1)


def reclassify_cmk(cmk):
    """
    INSAT-3DR CMK values:
        0 → unprocessed / fill  (invalid)
        1 → clear sky
        2 → probably cloudy
        3 → cloudy

    Output:
        0 → clear sky (or invalid)
        1 → cloud (values 2 or 3)
    """
    binary = np.zeros_like(cmk, dtype=np.uint8)
    binary[(cmk == 2) | (cmk == 3)] = 1
    return binary


# =====================================================
# PRECOMPUTE INDIA CROP INDICES (ONCE, from first file)
# =====================================================

h5_files = sorted(f for f in os.listdir(INPUT_DIR) if f.endswith(".h5"))
if not h5_files:
    raise FileNotFoundError(f"No .h5 files found in {INPUT_DIR}")

sample_file = h5_files[0]
print(f"Using '{sample_file}' to compute India bounding box …")

with h5py.File(os.path.join(INPUT_DIR, sample_file), "r") as f:
    lat, lon = read_latlon(f)
    ll_shape = lat.shape
    print(f"Lat/Lon grid shape       : {ll_shape}")

    # Diagnostic: print CMK dataset shape so any future dimension issues
    # are immediately obvious.
    cmk_raw_shape = f["CMK"].shape
    print(f"CMK dataset shape in HDF5: {cmk_raw_shape}")

    rmin, rmax, cmin, cmax = compute_india_bbox(lat, lon)
    print(f"India crop (lat/lon indices) : rows {rmin}:{rmax}, cols {cmin}:{cmax}")
    print(f"Crop size (at lat/lon res)   : {rmax-rmin+1} rows × {cmax-cmin+1} cols")

# =====================================================
# MAIN LOOP
# =====================================================
processed = skipped = 0

for fname in sorted(h5_files):
    print(f"\nProcessing: {fname}")
    fpath = os.path.join(INPUT_DIR, fname)

    try:
        with h5py.File(fpath, "r") as f:
            cmk, (r0, r1, c0, c1) = read_cmk_crop(f, rmin, rmax, cmin, cmax, ll_shape)
    except Exception as e:
        print(f"  ⚠  Could not read CMK: {e}")
        skipped += 1
        continue

    vals, cnts = np.unique(cmk, return_counts=True)
    print(f"  CMK shape : {cmk.shape}  |  values : {dict(zip(vals.tolist(), cnts.tolist()))}")

    valid_mask = (cmk == 1) | (cmk == 2) | (cmk == 3)
    if np.sum(valid_mask) == 0:
        print("  ⏭  Skipped (no valid CMK pixels)")
        skipped += 1
        continue

    binary_cmk = reclassify_cmk(cmk)
    frac = np.sum(binary_cmk[valid_mask] == 1) / np.sum(valid_mask)

    if frac < MIN_CLOUD_FRACTION:
        print(f"  ⏭  Skipped (cloud fraction = {frac:.2f} < {MIN_CLOUD_FRACTION})")
        skipped += 1
        continue

    # Build output filename from the HDF5 filename
    # Pattern: 3RIMG_<DDMONYYYY>_<HHMM>_L2B_CMK_...
    parts = fname.split("_")
    date_str = parts[1]          # e.g. 01OCT2025
    time_str = parts[2]          # e.g. 0015
    date_fmt = datetime.strptime(date_str, "%d%b%Y").strftime("%Y%m%d")
    out_name = f"binary_cmk_{date_fmt}_{time_str}.npy"
    out_path = os.path.join(OUTPUT_DIR, out_name)

    np.save(out_path, binary_cmk)
    processed += 1
    print(
        f"  ✅ Saved → {out_name} | "
        f"shape={binary_cmk.shape} | "
        f"cloud fraction={frac:.2f}"
    )

print(f"\n🎉 Done.  Saved: {processed}  Skipped: {skipped}")
```

```python
"""
predict_and_show.py
===================
Run cloud cover prediction from ONNX model and visualise results.

Dependencies:
    pip install onnxruntime numpy matplotlib

Notebook usage — just set the paths in the NOTEBOOK CONFIG section below.

CLI usage:
    # Single sequence (4 input files):
    python predict_and_show.py --model best_model.onnx \
        --files t1.npy t2.npy t3.npy t4.npy

    # Auto-discover sequences in a folder:
    python predict_and_show.py --model best_model.onnx \
        --dir /path/to/npy_folder --n_samples 3
"""

import os
import re
import sys
import argparse
import numpy as np
from glob import glob
from datetime import datetime
import cv2
from scipy.ndimage import gaussian_filter, binary_opening, binary_closing, label
# ── CONFIG (must match training) ──────────────────────────────────────────────
NATIVE_H = 758
NATIVE_W = 929
PAD_H    = int(__import__('math').ceil(NATIVE_H / 8)) * 8   # 760
PAD_W    = int(__import__('math').ceil(NATIVE_W / 8)) * 8   # 936
T_IN     = 4
T_OUT    = 4


# ─────────────────────────────────────────────────────────────────────────────
# PREPROCESSING
# ─────────────────────────────────────────────────────────────────────────────

def load_frame(path: str) -> np.ndarray:
    img = np.load(path).astype(np.float32)
    img = img[:NATIVE_H, :NATIVE_W]
    ph  = PAD_H - img.shape[0]
    pw  = PAD_W - img.shape[1]
    img = np.pad(img, ((0, ph), (0, pw)), mode='reflect')
    return img


def build_input(file_paths: list) -> np.ndarray:
    assert len(file_paths) == T_IN, f'Need {T_IN} files, got {len(file_paths)}'
    frames = np.stack([load_frame(p) for p in file_paths])   # (T, H, W)
    return frames[np.newaxis, ..., np.newaxis].astype(np.float32)  # (1,T,H,W,1)


# ─────────────────────────────────────────────────────────────────────────────
# ONNX INFERENCE
# ─────────────────────────────────────────────────────────────────────────────
# ─────────────────────────────────────────────
# OPTICAL FLOW HELPERS
# ─────────────────────────────────────────────

def compute_flow(prev_img, next_img):
    """
    Dense optical flow using Farneback.
    Input images must be float [0,1].
    """
    prev = (prev_img * 255).astype(np.uint8)
    nxt  = (next_img * 255).astype(np.uint8)

    flow = cv2.calcOpticalFlowFarneback(
        prev, nxt,
        None,
        pyr_scale=0.5,
        levels=3,
        winsize=25,
        iterations=3,
        poly_n=5,
        poly_sigma=1.2,
        flags=0
    )
    return flow


def warp_mask(mask, flow):
    """
    Warp binary mask using optical flow.
    """
    h, w = mask.shape

    grid_x, grid_y = np.meshgrid(np.arange(w), np.arange(h))

    map_x = (grid_x + flow[..., 0]).astype(np.float32)
    map_y = (grid_y + flow[..., 1]).astype(np.float32)

    warped = cv2.remap(
        mask.astype(np.float32),
        map_x,
        map_y,
        interpolation=cv2.INTER_LINEAR,
        borderMode=cv2.BORDER_REFLECT
    )

    return (warped > 0.5).astype(np.uint8)

def load_session(model_path: str, device: str = 'cpu'):
    try:
        import onnxruntime as ort
    except ImportError:
        sys.exit('ERROR: pip install onnxruntime')

    providers = (['CUDAExecutionProvider', 'CPUExecutionProvider']
                 if device == 'cuda' else ['CPUExecutionProvider'])

    opts = ort.SessionOptions()
    opts.graph_optimization_level = ort.GraphOptimizationLevel.ORT_ENABLE_ALL
    session = ort.InferenceSession(model_path, sess_options=opts, providers=providers)

    print(f'Model  : {model_path}')
    print(f'Input  : {session.get_inputs()[0].name}  {session.get_inputs()[0].shape}')
    print(f'Output : {session.get_outputs()[0].name}  {session.get_outputs()[0].shape}')
    print(f'Device : {session.get_providers()[0]}')
    return session


def run_inference(session, inp: np.ndarray) -> np.ndarray:
    """Returns (1, T_OUT, NATIVE_H, NATIVE_W, 1) float32."""
    name = session.get_inputs()[0].name
    out  = session.get_outputs()[0].name
    return session.run([out], {name: inp})[0]


# ─────────────────────────────────────────────────────────────────────────────
# VISUALISATION
# ─────────────────────────────────────────────────────────────────────────────



def visualise_and_save(input_files,
                       pred,
                       output_dir,
                       prefix,
                       threshold=0.5,
                       show=True):

    import matplotlib.pyplot as plt
    import matplotlib.gridspec as gridspec

    os.makedirs(output_dir, exist_ok=True)

    T = pred.shape[1]

    # Last two real inputs → motion estimation
    frame_prev = load_frame(input_files[-2])[:NATIVE_H, :NATIVE_W]
    frame_last = load_frame(input_files[-1])[:NATIVE_H, :NATIVE_W]

    flow = compute_flow(frame_prev, frame_last)

    probs = []
    masks = []

    prev_mask = None

    # ─────────────────────────────────────────
    # WEATHER + FLOW POSTPROCESSING
    # ─────────────────────────────────────────
    for t in range(T):

        prob = pred[0, t, :, :, 0]

        # 1️⃣ smooth probability
        prob = gaussian_filter(prob, sigma=1.2)

        # 2️⃣ adaptive threshold
        thr = np.percentile(prob, 65)
        mask = (prob >= thr).astype(np.uint8)

        # 3️⃣ morphology
        mask = binary_closing(mask, structure=np.ones((3,3)))
        mask = binary_opening(mask, structure=np.ones((2,2)))

        # 4️⃣ remove tiny blobs
        lbl, n = label(mask)
        clean = np.zeros_like(mask)
        for i in range(1, n+1):
            region = (lbl == i)
            if region.sum() > 40:
                clean[region] = 1
        mask = clean

        # 5️⃣ OPTICAL FLOW WARPING ⭐
        if prev_mask is not None:
            warped_prev = warp_mask(prev_mask, flow)

            # blend predicted + advected mask
            mask = ((0.6 * mask + 0.4 * warped_prev) > 0.5).astype(np.uint8)

        prev_mask = mask.copy()

        probs.append(prob)
        masks.append(mask)

        np.save(os.path.join(output_dir, f'{prefix}_prob_t{t+1}.npy'), prob)
        np.save(os.path.join(output_dir, f'{prefix}_mask_t{t+1}.npy'), mask)

    # ─────────────────────────────────────────
    # VISUALISATION (same layout)
    # ─────────────────────────────────────────
    ncols = T + 1
    fig = plt.figure(figsize=(ncols*3.2, 9), facecolor='#1a1a2e')

    gs = gridspec.GridSpec(3, ncols, hspace=0.35, wspace=0.05)

    def style(ax, title):
        ax.set_title(title, color='white', fontsize=8)
        ax.set_xticks([])
        ax.set_yticks([])

    # input
    ax = fig.add_subplot(gs[0,0])
    ax.imshow(frame_last, cmap='gray', vmin=0, vmax=1)
    style(ax, "Last input")

    ax = fig.add_subplot(gs[1,0])
    ax.imshow(frame_last, cmap='gray')
    style(ax, "")

    ax = fig.add_subplot(gs[2,0])
    ax.imshow((frame_last>0.5), cmap='gray')
    style(ax, "Input mask")

    # predictions
    for t in range(T):

        c = t+1

        ax = fig.add_subplot(gs[0,c])
        ax.imshow(probs[t], cmap='Blues', vmin=0, vmax=1)
        style(ax, f"Prob t+{t+1}")

        ax = fig.add_subplot(gs[1,c])
        ax.imshow(probs[t], cmap='inferno', vmin=0, vmax=1)
        style(ax, f"Detail t+{t+1}")

        ax = fig.add_subplot(gs[2,c])
        ax.imshow(masks[t], cmap='gray', vmin=0, vmax=1)
        style(ax, f"Mask t+{t+1} ({masks[t].mean()*100:.1f}%)")

    fig.suptitle(
        f"CrevNet Forecast · Optical-Flow Guided Binary Masks",
        color='white'
    )

    out_path = os.path.join(output_dir, f"{prefix}_prediction.png")
    plt.savefig(out_path, dpi=150, bbox_inches='tight',
                facecolor=fig.get_facecolor())

    print("Saved →", out_path)

    if show:
        plt.show()

    plt.close()
# ─────────────────────────────────────────────────────────────────────────────
# SEQUENCE DISCOVERY
# ─────────────────────────────────────────────────────────────────────────────

def parse_ts(fname):
    m = re.search(r'(\d{8})_(\d{4})', fname)
    return datetime.strptime(f'{m.group(1)}_{m.group(2)}', '%Y%m%d_%H%M') if m else None


def discover(folder, stride=4):
    files = sorted(glob(os.path.join(folder, '*.npy')))
    stamped = sorted([(parse_ts(os.path.basename(f)), f)
                      for f in files if parse_ts(os.path.basename(f))],
                     key=lambda x: x[0])
    paths = [f for _, f in stamped] if stamped else files
    return [paths[i:i+T_IN] for i in range(0, len(paths)-T_IN+1, stride)
            if len(paths[i:i+T_IN]) == T_IN]


# ─────────────────────────────────────────────────────────────────────────────
# MAIN
# ─────────────────────────────────────────────────────────────────────────────

def is_notebook():
    try:
        get_ipython(); return True
    except NameError:
        return False


def run(model_path, files=None, directory=None,
        output_dir='./predictions', threshold=0.5,
        device='cpu', stride=4, n_samples=None, show=True):

    session = load_session(model_path, device)

    # Build sequence list
    if files:
        sequences = [files];  prefixes = ['sequence_0000']
    elif directory:
        sequences = discover(directory, stride)
        if not sequences:
            sys.exit(f'ERROR: no sequences found in {directory}')
        if n_samples:
            import random; random.shuffle(sequences)
            sequences = sequences[:n_samples]
        prefixes = [f'sequence_{i:04d}' for i in range(len(sequences))]
        print(f'Found {len(sequences)} sequences in {directory}')
    else:
        sys.exit('ERROR: provide --files or --dir')

    os.makedirs(output_dir, exist_ok=True)
    print(f'\nOutput dir : {output_dir}')
    print(f'Threshold  : {threshold}')
    print(f'Sequences  : {len(sequences)}\n{"─"*50}')

    for i, (seq, prefix) in enumerate(zip(sequences, prefixes)):
        print(f'\n[{i+1}/{len(sequences)}] {prefix}')
        for j, f in enumerate(seq):
            print(f'  input t-{T_IN-j}: {os.path.basename(f)}')

        inp  = build_input(seq)
        pred = run_inference(session, inp)
        visualise_and_save(seq, pred, output_dir, prefix, threshold, show=show)

    print(f'\n{"─"*50}\nDone.  All outputs → {output_dir}')


if __name__ == '__main__' or is_notebook():

    if is_notebook():
        # ══════════════════════════════════════════════
        # NOTEBOOK CONFIG — edit these paths
        # ══════════════════════════════════════════════
        MODEL_PATH  = r"C:\Users\parth\OneDrive\Desktop\best_model.onnx"
        OUTPUT_DIR  = r"C:\Users\parth\OneDrive\Desktop\results"
        THRESHOLD   = 0.5
        DEVICE      = 'cpu'       # 'cpu' or 'cuda'
        SHOW_PLOTS  = True        # False on headless servers

        # Option A — single sequence: exactly 4 .npy files in time order
        INPUT_FILES = [
            r"C:\Users\parth\OneDrive\Desktop\dataset\INSET 3DR dataset\Dataset\cmk\binary_cmk_20250801_0015.npy",
            r"C:\Users\parth\OneDrive\Desktop\dataset\INSET 3DR dataset\Dataset\cmk\binary_cmk_20250801_0045.npy",
            r"C:\Users\parth\OneDrive\Desktop\dataset\INSET 3DR dataset\Dataset\cmk\binary_cmk_20250801_0115.npy",
            r"C:\Users\parth\OneDrive\Desktop\dataset\INSET 3DR dataset\Dataset\cmk\binary_cmk_20250801_0145.npy",
        ]
        INPUT_DIR   = None   # set to None when using INPUT_FILES

        # Option B — batch from folder (comment out A and uncomment B)
        # INPUT_FILES = None
        # INPUT_DIR   = '/kaggle/input/datasets/parthdhanker/cmk-cloud-mask/cmk'
        # N_SAMPLES   = 5   # how many random sequences to visualise
        N_SAMPLES = None
        # ══════════════════════════════════════════════

        run(model_path=MODEL_PATH,
            files=INPUT_FILES,
            directory=INPUT_DIR,
            output_dir=OUTPUT_DIR,
            threshold=THRESHOLD,
            device=DEVICE,
            n_samples=N_SAMPLES,
            show=SHOW_PLOTS)

    else:
        parser = argparse.ArgumentParser(description='CrevNet prediction + visualisation')
        parser.add_argument('--model',      required=True)
        parser.add_argument('--files',      nargs='+', default=None,
                            help=f'{T_IN} .npy input files in time order')
        parser.add_argument('--dir',        default=None,
                            help='Folder of .npy files (auto-sequenced)')
        parser.add_argument('--output_dir', default='./predictions')
        parser.add_argument('--threshold',  type=float, default=0.5)
        parser.add_argument('--device',     default='cpu', choices=['cpu','cuda'])
        parser.add_argument('--stride',     type=int, default=4)
        parser.add_argument('--n_samples',  type=int, default=None,
                            help='Max sequences to process (--dir mode)')
        parser.add_argument('--no_show',    action='store_true',
                            help='Do not display plots (save only)')
        args = parser.parse_args()

        run(model_path=args.model,
            files=args.files,
            directory=args.dir,
            output_dir=args.output_dir,
            threshold=args.threshold,
            device=args.device,
            stride=args.stride,
            n_samples=args.n_samples,
            show=not args.no_show)
```
